```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv.imread('7687.jpg', cv.IMREAD_COLOR)
```

```python
plt.imshow(img)
```

```python
#    
gray_img = img.copy()
gray_img = cv.cvtColor(gray_img, cv.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray_img)
```

```python
#     -     
blurred = cv.GaussianBlur(gray_img, (3, 3), 0)
sharp = cv.addWeighted(gray_img, 1.2, blurred, -0.9 , 0)
plt.imshow(sharp)
```

```python
#  ,     
blurred_2 = cv.GaussianBlur(gray_img, (9, 9), 10)
sharp_2 = cv.addWeighted(gray_img, 1.8, blurred_2, -0.9, 0)
plt.imshow(sharp_2)
```
